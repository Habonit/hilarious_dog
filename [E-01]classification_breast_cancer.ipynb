```python
from sklearn.datasets import load_wine
from sklearn.datasets import load_digits
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
breast_cancer = load_breast_cancer()
breast_cancer_data = breast_cancer.data
breast_cancer_feature = breast_cancer.feature_names
breast_cancer_label = breast_cancer.target
breast_cancer_df = pd.DataFrame(breast_cancer_data, columns = breast_cancer_feature)
```

```python
print(breast_cancer_label)
print(breast_cancer.DESCR)
```

```python
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data, 
                                                    breast_cancer_label, 
                                                    test_size = 0.2,
                                                    random_state = 42)
```

```python
print("Breast_Cancer_data")
print("-"*75)

decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print("Decision Tree 모델 학습", "\n\n")
print(classification_report(y_test, y_pred))
print("-"*75)

random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print("Random Forest 모델 학습", "\n\n")
print(classification_report(y_test, y_pred))
print("-"*75)

svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print("SVM 모델 학습", "\n\n")
print(classification_report(y_test, y_pred))
print("-"*75)

sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print("SGD 모델 학습", "\n\n")
print(classification_report(y_test, y_pred))
print("-"*75)

logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print("Logistic Regression 모델 학습", "\n\n")
print(classification_report(y_test, y_pred))
print("-"*75)
```

해당 자료는 SVM 모델을 사용하는 것이 현명합니다.

SVM 모델을 사용했을 때, 음성(1)의 recall율이 가장 높았기 때문입니다. 

음성(1)의 recall율이 중요한 이유는, 유방암이 아니라고 판정했을 때, 그 판정이 거짓인 경우가 가장 문제적인 상황이기 때문입니다.

따라서, 음성 판정이 거짓일 확률이 가장 낮은 모델인 SVM과 SGD 중에서 나머지 통계에서 우위를 보이는 SVM 모델을 사용하는 것이 타당합니다.
