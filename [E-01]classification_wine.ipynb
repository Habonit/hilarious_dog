```python
from sklearn.datasets import load_wine
from sklearn.datasets import load_digits
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
wine = load_wine()
wine_data = wine.data
wine_feature = wine.feature_names
wine_label = wine.target
wine_df = pd.DataFrame(wine_data, columns = wine_feature)
```

```python
print(wine_label)
print(wine.DESCR)
```

```python
X_train, X_test, y_train, y_test = train_test_split(wine_data, 
                                                    wine_label, 
                                                    test_size = 0.2,
                                                    random_state = 42)
```

```python
print("Wine_data")
print("-"*75)

decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print("Decision Tree 모델 학습", "\n\n")
print(classification_report(y_test, y_pred))
print("-"*75)

random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print("Random Forest 모델 학습", "\n\n")
print(classification_report(y_test, y_pred))
print("-"*75)

svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print("SVM 모델 학습", "\n\n")
print(classification_report(y_test, y_pred))
print("-"*75)

sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print("SGD 모델 학습", "\n\n")
print(classification_report(y_test, y_pred))
print("-"*75)

logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print("Logistic Regression 모델 학습", "\n\n")
print(classification_report(y_test, y_pred))
print("-"*75)
```

```python
해당 자료는 RandomForest 모델을 사용하는 것이 현명합니다.

RandomForest 모델을 사용했을 때, 전체적인 precision과 recall, f1-score가 가장 높았기 때문입니다.  
```
