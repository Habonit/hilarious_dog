## **1.1. 필요한 모듈을 불러와 유방암 데이터를 준비하기**

```python
from sklearn.datasets import load_breast_cancer 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

breast_cancer = load_breast_cancer()
breast_cancer_data = breast_cancer.data
breast_cancer_feature = breast_cancer.feature_names
breast_cancer_label = breast_cancer.target
breast_cancer_df = pd.DataFrame(breast_cancer_data, columns = breast_cancer_feature)
```

## **1.2. train, test 데이터 분리하기**

```python
from sklearn.model_selection import train_test_split
# 질문1: train_test_split을 사용하지 않는다면, 임의 추출할 수 있는 다른 방법은 무엇이 있을까?
# 질문2: 최적의 test_size는 몇인가? 너무 많거나 너무 적으면 어떤 오류가 만들어질까?
# 질문3: 이상적인 random_state는 존재하는가?
# 질문4: shuffle, stratify 인자의 역할은 각각 무엇인가?

X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data, 
                                                    breast_cancer_label, 
                                                    test_size = 0.2,
                                                    random_state = 42)
```

    2) 질문2: 최적의 test_size는 몇인가? 너무 많거나 너무 적으면 어떤 오류가 만들어질까?
    3) 질문3: 이상적인 random_state는 존재하는가?
    4) 질문4: shuffle, stratify 인자의 역할은 각각 무엇인가?

#질문 1: SVM, SGD, Logistic Regression의 유도 공식은 어떻게 되는가?

### 1.3.1. DecisionTree 모델

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

# 질문1 : DecisionTree에서 random_state의 역할은 어떻게 되는가?
# 질문2 : weighted avg의 가중치는 어떻게 되는가?

decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print("1. Decision Tree 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 1.3.2. RandomForest 모델

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

#질문1 : RandomForest에서 random_state의 역할은 무엇인가?

random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print("1. Random Forest 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3.  classification_report", classification_report(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 1.3.3. SVM 모델

```python
from sklearn import svm
from sklearn.metrics import classification_report

svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print("1. SVM 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred),sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 1.3.4. SGD 모델

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import classification_report

sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print("1. SGD 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 1.3.5 LogisticRegression 모델

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print("1. Logistic Regression 모델 학습",  sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred),  sep = "\n"*2, end="\n"*2)
print("-"*75)
```

**remark1**<br>
: 위와 같이 실행을 해보면, 5번 logistic regression 모델에서 iteration이 max_iter 값을 초과했다는 오류가 나옵니다. max_iteration을 default 값인 100에서 4000정도로 올려주면 해결이 됩니다.

아직 logistic regression이 어떤 수리적인 공식을 통해 유도되는지 파악하지 못하였기 때문에, max_iteration의 횟수에 따른 오류가 왜 발생하는지는 분석하기 어렵습니다. 이는 추후 공부해볼 내용으로 남겨놓겠습니다. 

```python
logistic_model = LogisticRegression(max_iter = 4000)
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print("1. max_iter 값을 4000으로 수정한 Logistic Regression 모델",sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred),  sep = "\n"*2, end="\n"*2)
print("-"*75)
```

**remark2. classification_report 용어 정리**

    1) precision(정밀도) : confusion matrix 상에서 TP / (TP + FP)

    2) recall(재현율) : confusion matrix 상에서 TP / (FN + TP)

    3) f1-score: precision과 recall의 harmonic average / 높을수록 좋은 모델이다. 

    4) accuracy(정확도) : confusion matrix 상에서 TP + TN / (TP + TN + FP + FN)

## 1.4. 결론

해당 자료는 SVM 모델을 사용하는 것이 현명합니다.

SVM 모델을 사용했을 때, 음성(1)의 recall율이 가장 높았기 때문입니다. 

breast_cancer의 target_names를 확인해보면 malignant(양성)이 0, benign(음성)이 1로 확인됩니다. 

음성의 recall율이 중요한 이유는, 유방암이 아니라고 판정했을 때, 그 판정이 거짓인 경우가 가장 문제적인 상황입니다. 즉, FN이 낮아야 좋은 통계라 할 수 있습니다.

반면에, 양성이라고 판정한 후 추후에 확인해보니 음성이었다면, 이는 위의 경우보다 문제가 심각하지 않습니다. 즉, FP가 높은 것이 그리 심각한 상황은 아닙니다. 

따라서, 음성 판정이 거짓일 확률이 가장 낮은 모델인 SVM과 SGD 중에서 나머지 통계에서 우위를 보이는 SVM 모델을 사용하는 것이 타당합니다.
