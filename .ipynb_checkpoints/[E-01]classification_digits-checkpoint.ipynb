## **2.1. 필요한 모듈을 불러와 손글씨 데이터를 준비하기**

```python
from sklearn.datasets import load_digits
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
from sklearn.datasets import load_digits
import pandas as pd

digits = load_digits()
digits_data = digits.data
digits_feature = digits.feature_names
digits_label = digits.target
digits_df = pd.DataFrame(digits_data, columns = digits_feature)
```

```python
print(digits_label)
print(digits.DESCR)
```

```python
digits_df
```

***

## **2.2. train, test 데이터 분리하기**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size = 0.2,
                                                    random_state = 42)
```

***

## 2.3. 다양한 모델로 학습시켜보기
    
    대상 모델
    1) DecisionTree 모델
    2) RandomForeset 모델
    3) SVM 모델
    4) SGD 모델
    5) Logistic Regression 모델
    
***

### 2.3.1. DecisionTree 모델

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print("1. Decision Tree 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 2.3.2. RandomForest 모델

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print("1. Random Forest 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3.  classification_report", classification_report(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 2.3.3. SVM 모델

```python
from sklearn import svm
from sklearn.metrics import classification_report

svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print("1. SVM 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred),sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 2.3.4. SGD 모델

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import classification_report

sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print("1. SGD 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 2.3.5. LogisticRegression 모델

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

logistic_model = LogisticRegression(max_iter = 4000)
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print("1. Logistic Regression 모델 학습",  sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred),  sep = "\n"*2, end="\n"*2)
print("-"*75)
```

## 2.4. 결론

해당 자료는 SVM 모델을 사용하는 것이 현명합니다.

SVM 모델을 사용했을 때, 전체적인 precision과 recall, f1-score가 가장 높았기 때문입니다.
