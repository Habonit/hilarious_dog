## **3.1. 필요한 모듈을 불러와 손글씨 데이터를 준비하기**

```python
from sklearn.datasets import load_wine
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine()
wine_data = wine.data
wine_feature = wine.feature_names
wine_label = wine.target
wine_df = pd.DataFrame(wine_data, columns = wine_feature)
```

```python
print(wine_label)
print(wine.DESCR)
```

```python
wine_df
```

***

## **3.2. train, test 데이터 분리하기**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(wine_data, 
                                                    wine_label, 
                                                    test_size = 0.2,
                                                    random_state = 42)
```

***
## 3.3. 다양한 모델로 학습시켜보기
    
    대상 모델
    1) DecisionTree 모델
    2) RandomForeset 모델
    3) SVM 모델
    4) SGD 모델
    5) Logistic Regression 모델
    
***

### 3.3.1. DecisionTree 모델

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print("1. Decision Tree 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 3.3.2. RandomForest 모델

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print("1. Random Forest 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3.  classification_report", classification_report(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 3.3.3. SVM 모델

```python
from sklearn import svm
from sklearn.metrics import classification_report

svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print("1. SVM 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred),sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 3.3.4. SGD 모델

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import classification_report

sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print("1. SGD 모델 학습", sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
```

### 3.3.5. LogisticRegression 모델

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

logistic_model = LogisticRegression(max_iter = 4000)
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print("1. Logistic Regression 모델 학습",  sep = "\n"*2, end="\n"*2)
print("-"*75)
print("2. confusion_matrix", confusion_matrix(y_test, y_pred), sep = "\n"*2, end="\n"*2)
print("-"*75)
print("3. classification_report", classification_report(y_test, y_pred),  sep = "\n"*2, end="\n"*2)
print("-"*75)
```

## 3.4. 결론

해당 자료는 RandomForest나 Logistic Regression 모델을 사용하는 것이 현명합니다.

RandomForest 모델을 사용했을 때, 전체적인 precision과 recall, f1-score가 가장 높았기 때문입니다.  

단, 유방암 데이터나 손글씨 데이터에 비해 전체적인 데이터 수가 적기 때문에 모델을 평가하기에 부적합할 수 있다고 생각합니다.

깊이 있는 공부가 부족했던 것을 팀원과의 프로젝트 회고를 통해 보충할 수 있었다. 제출하기 직전인 일요일, 스터디 팀원과 프로젝트에 대한 자신의 관점을 공유하고 상호 피드백을 주고 받는 시간을 가졌다. 그 과정에서 알게 된 사실이 매우 많았다.

두 번째: 첫 번째 사실의 연장선에서 마크 다운 문법을 잘 숙지하여 가독성 있게 프로젝트를 작성하는 것이 필요함을 알 수 있었다.
